```python
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
```

```python
# load the excel data

df = pd.read_excel('Oil well.xlsx',header=2)
df.head()
```

```python
# get the shape of the data

df.shape
```

```python
# rename the columns

col_mapper = {'\nOil volume (m3/day)':'Oil_volume',
             'Volume of liquid (m3/day)':'Volume_of_liquid',
             '\nGas volume (m3/day)':'Gas_volume',
             'Water volume (m3/day)':'Water_volume',
             'Water cut (%)':'Water_cut',
             '\nWorking hours':'Working_hours',
             '\nDynamic level (m)':'Dynamic_level',
             '\nReservoir pressure (atm)':'Reservoir_pressure'}

df.rename(col_mapper,axis=1,inplace=True)
```

```python
df.head()
```

```python
# remove the % from the water cut column

df.rename({'Water cut (%) ':'Water_cut'},axis=1,inplace=True)
```

```python
# make all the column names as lowercase

df.columns = df.columns.str.lower()
```

```python
# check for missing values in the data
df.isna().sum()
```

**There are no missing values in the data**

```python
# check the info of the data
df.info()
```

```python
# generate the month column 

df['month'] = df['date'].dt.month_name()
```

```python
# generate the year column

df['year'] = df['date'].dt.year
```

```python
# set the date column as the index

df.set_index('date',inplace=True)
```

```python
df.head()
```

```python
# check the date range in the data

min_range = df.index.min()
max_range = df.index.max()

min_range,max_range
```

```python
# unique values in the month column

df['month'].unique()
```

```python
# unique values in the year column
df['year'].unique()
```

```python
# check the time series of all columns in the data

plt.figure(figsize=(15,10))

for i,col in enumerate(df.columns[0:8]):
    
    plt.subplot(4,2,i+1)
    df[col].plot()
    plt.title(col)
    plt.tight_layout()
    
```

```python
# plot the trends in the chart for a particular year



df_2020 = df[df['year'] == 2020]
df_2020

# check the time series of all columns in the data
plt.figure(figsize=(15,10))

for i,col in enumerate(df_2020.columns[0:8]):
    
    plt.subplot(4,2,i+1)
    df_2020[col].plot()
    plt.title(col)
    plt.tight_layout()
```
